## Q4.
### Run Apriori algorithm to find frequent itemsets and association rules<br>
### 1.1 Use minimum support as 50% and minimum confidence as 75%<br>
### 1.2 Use minimum support as 60% and minimum confidence as 60 %<br>

```python
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("Market_Basket_Optimisation.csv")
df.head()
```

```python
df.fillna('!',inplace=True)
trans = df.values.tolist()
```

```python
for i in range(len(trans)):
    trans[i] = [x for x in trans[i] if not x=='!']
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(trans).transform(trans)
transactions = pd.DataFrame(te_ary, columns=te.columns_)
transactions
```

# 1.1 Use minimum support as 50% and minimum confidence as 75%

```python
freq_items = apriori(transactions, min_support=0.05, use_colnames=True, verbose=1)
freq_items.head(10)
```

```python
freq_items['length'] = freq_items['itemsets'].apply(lambda x: len(x))
```

```python
freq_items.sort_values('length',ascending=False)
```

```python
rules = association_rules(freq_items, metric="confidence", min_threshold=0.075)
rules.head()
```

```python
rules.sort_values('confidence', ascending=False)
```

# 1.1 Use minimum support as 60% and minimum confidence as 60%

```python
freq_items = apriori(transactions, min_support=0.006, use_colnames=True, verbose=1)
freq_items.head(10)
```

```python
freq_items['length'] = freq_items['itemsets'].apply(lambda x: len(x))
```

```python
freq_items.sort_values('length',ascending=False)
```

```python
rules = association_rules(freq_items, metric="confidence", min_threshold=0.006)
rules.head()
```

```python
rules.sort_values('confidence', ascending=False)
```
